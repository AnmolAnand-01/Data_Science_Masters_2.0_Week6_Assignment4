Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a micro web framework for python. It is designed to be lightweight, simple and easy to use framework making it popular among the developers fo developing web applications and APIs.
Advantages of flask:-
1. Flask is built with python in mind and follows pythonic principles which means it is consistent with python's design philosophy. Anyone having knowledge of python will find flask easy to use.
2. Flask includes a built-in development server, making it easy to get started with development work without the need of additional setup.
3. Flask is well suited for building RESTful APIs due to simplicity and ease with which it can handle HTTP requests and responses.
4. Flask's simplicity makes it easy to write unit tests and perform testing of the application, ensuring it's reliability and stability.
5. Flask provides a wide range of extensions and third-pary packages that can be easily integrated into our application to add functionalities like authentication, database support and more.

Q3. What is App routing in Flask? Why do we use app routes?

In FLask, app routing refers to the process of defining the URL routes for the web applications. These routes determine how the application responds to different URLs and HTTP methods(eg:- GET, PUT, POST, DELETE) by mapping them to specific functions in our code. 
1. We specify routes of our application by using app.route() decorator provided by flask.
2. We can also define rotutes with dynamic components by specifying variabl parts in the URL. These values are enclosed within '<' and '>'.
3. Routes map specific URLs to corresponding view functions, making it easy to navigate and interact with our web applications.
4. Routes allows us to organise our code into smaller self contained functions that handles specific tasks.
5. We can use various HTTP methods(eg:- GET, PUT, POST, DELETE) to specify how route should respond to various types of requests.
6. By using dynamic routes with variables, we can create routes that responds to user inputs, enabling our application to display personalised content or process user submitted data.

Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def details():
    return """<pre>Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-999900 </pre>"""

@app.route("/welcome")
def message():
    return "Welcome to ABC Corporation"

if __name__ == "__main__":
    app.run(host="0.0.0.0")
```
